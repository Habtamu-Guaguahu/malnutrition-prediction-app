```python
######################################################################################
# ===================================================================================#
# Machine Learning Algorithm For Predicting Acute Manutrition in Sub-Saharan Africa  #
# ===================================================================================#
######################################################################################

# ==============================================
# 1. COMPLETE IMPORTS & SETUP
# ==============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, cross_val_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import (
    SelectKBest, 
    SelectFromModel,
    RFE,
    mutual_info_classif,
    chi2,
    VarianceThreshold
)
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# Models
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

# Imbalance
from imblearn.over_sampling import SMOTEN
from imblearn.pipeline import Pipeline as ImbPipeline

# Metrics
from sklearn.metrics import (
    roc_auc_score, roc_curve, 
    confusion_matrix, classification_report,
    precision_recall_curve, average_precision_score
)

# Interpretation
import shap

# Warnings
import warnings
warnings.filterwarnings('ignore')

# Set random seed
np.random.seed(42)

print("✅ All libraries imported successfully!")
```

```python

# ==============================================
# 1. LOAD AND PREPARE DATA
# ==============================================

# Load your data
data = pd.read_stata(r"C:\Users\hp\Documents\Materials\Machine Learning\data\imputeddata3.dta")

# Encode target
data['wasting'] = data['wasting'].map({'No': 0, 'Yes': 1})
target = 'wasting'

# Split features and target
y = data[target]
X = data.drop(columns=[target])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=42
)

print("\n📊 DATA LOADED:")
print(f"   Train set: {X_train.shape[0]:,} samples")
print(f"   Test set: {X_test.shape[0]:,} samples")
print(f"   Original features: {X_train.shape[1]}")
```

```python
# ===============================================
# 2.APPLY SMOTEN ON RAW TRAIN DATA
# ===============================================

smote = SMOTEN(random_state=42, k_neighbors=3)

X_train_bal_raw, y_train_bal = smote.fit_resample(X_train, y_train)

print("\n⚖️ SMOTEN APPLIED (RAW DATA):")
print(f"   Before: {y_train.value_counts().to_dict()}")
print(f"   After: {pd.Series(y_train_bal).value_counts().to_dict()}")
```

```python

# ==============================================
# 3. IDENTIFY COLUMN TYPES
# ==============================================

categorical_cols = X_train.select_dtypes(include=['object', 'category']).columns.tolist()
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()

print(f"\n📋 FEATURE TYPES:")
print(f"   Categorical: {len(categorical_cols)} columns")
print(f"   Numeric: {len(numeric_cols)} columns")
```

```python
# ==============================================
# 4. SMART PREPROCESSOR
#    - Binary categorical → keep as 0/1
#    - Multi-category → full one-hot encoding
# ==============================================

from sklearn.preprocessing import OneHotEncoder

# Identify categorical variables
categorical_cols = X_train.select_dtypes(include=['object', 'category']).columns.tolist()

# Separate binary and multi-category categorical variables
binary_cols = [col for col in categorical_cols if X_train[col].nunique() == 2]
multi_cat_cols = [col for col in categorical_cols if X_train[col].nunique() > 2]

print(f"Binary categorical variables: {len(binary_cols)}")
print(f"Multi-category variables: {len(multi_cat_cols)}")

# Convert binary categories to 0/1 using LabelEncoder-like mapping
for col in binary_cols:
    X_train_bal_raw[col] = X_train_bal_raw[col].astype('category').cat.codes
    X_test[col] = X_test[col].astype('category').cat.codes

# Create preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('multi_cat', OneHotEncoder(
            handle_unknown='ignore',
            sparse_output=False
        ), multi_cat_cols),
        ('num', StandardScaler(), numeric_cols)
    ],
    remainder='passthrough',   # keep binary columns as they are
    verbose_feature_names_out=False
)

# Fit and transform
X_train_enc = preprocessor.fit_transform(X_train_bal_raw)
X_test_enc = preprocessor.transform(X_test)

# Get feature names properly
feature_names = preprocessor.get_feature_names_out()

# Convert to DataFrame
X_train_enc = pd.DataFrame(X_train_enc, columns=feature_names)
X_test_enc = pd.DataFrame(X_test_enc, columns=feature_names)

print("\n✅ SMART ENCODING COMPLETE")
print(f"Encoded features: {X_train_enc.shape[1]}")
```

```python
# ==============================================
# 5. APPLY MULTIPLE FEATURE SELECTION METHODS
# ==============================================

print("\n" + "="*80)
print("🔍 APPLYING FEATURE SELECTION METHODS")
print("="*80)

feature_selection_results = {}

X_fs = X_train_enc.copy()
y_fs = y_train_bal.copy()

# ------------------------------------------------
# 5.1 Variance Threshold
# ------------------------------------------------
print("\n📌 Variance Threshold...")

selector_var = VarianceThreshold(threshold=0.01)
selector_var.fit(X_fs)

var_features = X_fs.columns[selector_var.get_support()]

feature_selection_results['variance_threshold'] = {
    'features': list(var_features),
    'n_features': len(var_features)
}

print(f"   ✅ Selected {len(var_features)} features")


# ------------------------------------------------
# 5.2 Mutual Information
# ------------------------------------------------
print("\n📌 Mutual Information...")

n_features_mi = min(50, X_fs.shape[1])

selector_mi = SelectKBest(
    score_func=mutual_info_classif,
    k=n_features_mi
)

selector_mi.fit(X_fs, y_fs)

mi_features = X_fs.columns[selector_mi.get_support()]

feature_selection_results['mutual_info'] = {
    'features': list(mi_features),
    'n_features': len(mi_features)
}

print(f"   ✅ Selected {len(mi_features)} features")


# ------------------------------------------------
# 5.3 Chi-Square
# IMPORTANT: Chi2 requires NON-NEGATIVE features
# Do NOT use scaled variables
# ------------------------------------------------
print("\n📌 Chi-Square...")

# Create copy
X_chi2 = X_fs.copy()

# Ensure no negative values (due to scaling)
X_chi2 = X_chi2 - X_chi2.min()

n_features_chi2 = min(50, X_fs.shape[1])

selector_chi2 = SelectKBest(
    score_func=chi2,
    k=n_features_chi2
)

selector_chi2.fit(X_chi2, y_fs)

chi2_features = X_fs.columns[selector_chi2.get_support()]

feature_selection_results['chi2'] = {
    'features': list(chi2_features),
    'n_features': len(chi2_features)
}

print(f"   ✅ Selected {len(chi2_features)} features")


# ------------------------------------------------
# 5.4 SelectFromModel (Random Forest)
# ------------------------------------------------
print("\n📌 SelectFromModel (Random Forest)...")

rf_selector = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    n_jobs=-1
)

rf_selector.fit(X_fs, y_fs)

selector_rf = SelectFromModel(
    rf_selector,
    threshold='median',
    prefit=True
)

rf_features = X_fs.columns[selector_rf.get_support()]

feature_selection_results['select_from_model'] = {
    'features': list(rf_features),
    'n_features': len(rf_features)
}

print(f"   ✅ Selected {len(rf_features)} features")


# ------------------------------------------------
# 5.5 Recursive Feature Elimination (RFE)
# ------------------------------------------------
print("\n📌 Recursive Feature Elimination...")

n_features_rfe = min(40, X_fs.shape[1])

estimator = LogisticRegression(
    max_iter=3000,
    random_state=42,
    solver='liblinear'
)

selector_rfe = RFE(
    estimator,
    n_features_to_select=n_features_rfe,
    step=5
)

selector_rfe.fit(X_fs, y_fs)

rfe_features = X_fs.columns[selector_rfe.get_support()]

feature_selection_results['rfe'] = {
    'features': list(rfe_features),
    'n_features': len(rfe_features)
}

print(f"   ✅ Selected {len(rfe_features)} features")


print("\n✅ ALL FEATURE SELECTION METHODS COMPLETED")
```

```python
# ==============================================
# 6. CREATE CONSENSUS FEATURE SET (SELECTED BY ≥2 METHODS)
# ==============================================

print("\n" + "="*80)
print("🎯 CREATING CONSENSUS FEATURE SET (≥2 METHODS)")
print("="*80)

# ------------------------------------------------
# Step 1: Count votes for each feature
# ------------------------------------------------
feature_votes = {}

for method, result in feature_selection_results.items():
    features = result['features']
    
    for feature in features:
        feature_votes[feature] = feature_votes.get(feature, 0) + 1


# ------------------------------------------------
# Step 2: Vote distribution summary
# ------------------------------------------------
print("\n📊 VOTE DISTRIBUTION:")

vote_distribution = {}

for votes in feature_votes.values():
    vote_distribution[votes] = vote_distribution.get(votes, 0) + 1

total_features_considered = len(feature_votes)

for votes in sorted(vote_distribution.keys()):
    count = vote_distribution[votes]
    percentage = (count / total_features_considered) * 100
    bar = '█' * int(percentage / 4)
    print(f"   {votes} method(s): {count:5d} features ({percentage:5.1f}%) {bar}")


# ------------------------------------------------
# Step 3: Create Consensus Set (≥2 votes)
# ------------------------------------------------
MIN_VOTES = 2

consensus_features = [
    feature for feature, votes in feature_votes.items()
    if votes >= MIN_VOTES
]

# Sort consensus features by number of votes (descending)
consensus_features_sorted = sorted(
    consensus_features,
    key=lambda x: feature_votes[x],
    reverse=True
)

print("\n✅ CONSENSUS FEATURE SET SUMMARY:")
print(f"   Minimum votes required: {MIN_VOTES}")
print(f"   Total features in consensus set: {len(consensus_features_sorted):,}")
print(f"   Original encoded features: {X_train_enc.shape[1]:,}")

reduction = (1 - len(consensus_features_sorted)/X_train_enc.shape[1]) * 100
print(f"   Dimensionality reduction: {reduction:.1f}%")


# ------------------------------------------------
# Step 4: Display Top Consensus Features
# ------------------------------------------------
print("\n🏆 TOP CONSENSUS FEATURES:")

top_k = 20
top_consensus = sorted(
    feature_votes.items(),
    key=lambda x: x[1],
    reverse=True
)[:top_k]

for i, (feature, votes) in enumerate(top_consensus, 1):
    short_feature = feature[:60] + '...' if len(feature) > 60 else feature
    print(f"   {i:2d}. {short_feature:<60} (voted by {votes} methods)")


# ------------------------------------------------
# Step 5: Create Final Training/Test Sets
# ------------------------------------------------
X_train_consensus_df = X_train_enc[consensus_features_sorted].copy()
X_test_consensus_df = X_test_enc[consensus_features_sorted].copy()

print("\n✅ CONSENSUS DATASETS CREATED")
print(f"   Train shape: {X_train_consensus_df.shape}")
print(f"   Test shape:  {X_test_consensus_df.shape}")
```

```python
# ==============================================
# 7. CREATE FINAL CONSENSUS DATASET
# ==============================================

print("\n" + "="*80)
print("📦 BUILDING FINAL CONSENSUS DATASET")
print("="*80)

# ------------------------------------------------
# Safety Check: Ensure consensus_features exists
# ------------------------------------------------
if 'consensus_features_sorted' in globals():
    selected_features = consensus_features_sorted
else:
    selected_features = consensus_features

# ------------------------------------------------
# Keep only features that truly exist
# ------------------------------------------------
valid_consensus_features = [
    f for f in selected_features
    if f in X_train_enc.columns
]

# ------------------------------------------------
# Safety check: Prevent empty feature set
# ------------------------------------------------
if len(valid_consensus_features) == 0:
    raise ValueError("❌ No valid consensus features found. Check feature selection steps.")

# ------------------------------------------------
# Create final datasets
# ------------------------------------------------
X_train_consensus_df = X_train_enc[valid_consensus_features].copy()
X_test_consensus_df = X_test_enc[valid_consensus_features].copy()

# ------------------------------------------------
# Final validation checks
# ------------------------------------------------
assert list(X_train_consensus_df.columns) == list(X_test_consensus_df.columns), \
    "❌ Train and test feature mismatch!"

print("\n✅ CONSENSUS DATASET SUCCESSFULLY CREATED")
print(f"   Selected features: {len(valid_consensus_features)}")
print(f"   Train shape: {X_train_consensus_df.shape}")
print(f"   Test shape:  {X_test_consensus_df.shape}")

reduction_percent = (
    1 - len(valid_consensus_features) / X_train_enc.shape[1]
) * 100

print(f"   Dimensionality reduction: {reduction_percent:.1f}%")
```

```python
# ==============================================
# 8.1 LOGISTIC REGRESSION
# ==============================================

print("\n" + "="*80)
print("🚀 TRAINING: LOGISTIC REGRESSION")
print("="*80)

# Initialize model
log_model = LogisticRegression(
    max_iter=2000,
    solver='liblinear',
    random_state=42
)

# Fit model on balanced training data
log_model.fit(X_train_consensus_df, y_train_bal)

# ==============================================
# Logistic Regression with Youden Index Cutoff
# ==============================================

print("\n" + "="*80)
print("🚀 LOGISTIC REGRESSION (Youden Index Threshold)")
print("="*80)

# Predict probabilities
y_pred_proba_log = log_model.predict_proba(X_test_consensus_df)[:, 1]

# =====================================
# 1️⃣ Compute ROC Curve
# =====================================
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba_log)

# =====================================
# 2️⃣ Compute Youden Index
# =====================================
youden_index = tpr - fpr
optimal_idx = np.argmax(youden_index)
optimal_threshold = thresholds[optimal_idx]

print(f"\n🎯 Optimal Threshold (Youden Index): {optimal_threshold:.4f}")

# =====================================
# 3️⃣ Classify using optimal threshold
# =====================================
y_pred_log = (y_pred_proba_log >= optimal_threshold).astype(int)

# =====================================
# 4️⃣ Evaluation
# =====================================

# ROC-AUC
auc_log = roc_auc_score(y_test, y_pred_proba_log)
print(f"\n📊 ROC-AUC Score: {auc_log:.4f}")

# PR-AUC
pr_auc_log = average_precision_score(y_test, y_pred_proba_log)
print(f"📊 PR-AUC Score: {pr_auc_log:.4f}")

# Confusion Matrix
cm_log = confusion_matrix(y_test, y_pred_log)
print("\n📌 Confusion Matrix (Youden threshold):")
print(cm_log)

# Classification Report
print("\n📌 Classification Report:")
print(classification_report(y_test, y_pred_log, digits=4))

# =====================================
# 5️⃣ Extract Sensitivity & Specificity
# =====================================
tn, fp, fn, tp = cm_log.ravel()

sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)

print(f"\n📊 Sensitivity (Recall+): {sensitivity:.4f}")
print(f"📊 Specificity (Recall-): {specificity:.4f}")
print(f"📊 Youden Index: {sensitivity + specificity - 1:.4f}")


plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f'Logistic (AUC = {auc_log:.3f})')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Logistic Regression')
plt.legend()
plt.show()
```

```python
# ==============================================
# 📊 MODEL PERFORMANCE (NO RETRAINING)
# ==============================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import (
    roc_auc_score, roc_curve,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)


# ==========================================
# 📦 Collect All Model Probabilities
# ==========================================

models_proba = {
    "Logistic Regression": y_pred_proba_log,
    "Random Forest": y_pred_proba_rf,
    "LightGBM": y_pred_proba_lgb,
    "Decision Tree": y_pred_proba_dt,
    "Naive Bayes": y_pred_proba_gnb,
    "XGBoost": y_pred_proba_xgb,
    "CatBoost": y_pred_proba_cat,
    "SGD": y_pred_proba_sgd,
    "SVM": y_pred_proba_svm,
    "KNN": y_pred_proba_knn
}

from sklearn.metrics import roc_auc_score
import numpy as np

def auc_ci_bootstrap(y_true, y_proba, n_bootstraps=1000):

    rng = np.random.RandomState(42)
    bootstrapped_scores = []

    y_true = np.array(y_true)
    y_proba = np.array(y_proba)

    for i in range(n_bootstraps):
        indices = rng.randint(0, len(y_proba), len(y_proba))

        if len(np.unique(y_true[indices])) < 2:
            continue

        score = roc_auc_score(y_true[indices], y_proba[indices])
        bootstrapped_scores.append(score)

    lower = np.percentile(bootstrapped_scores, 2.5)
    upper = np.percentile(bootstrapped_scores, 97.5)

    return lower, upper

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_curve, confusion_matrix
)

results = []

for name, y_proba in models_proba.items():

    # ==============================
    # 1️⃣ AUC + 95% CI
    # ==============================
    auc = roc_auc_score(y_test, y_proba)
    lower_ci, upper_ci = auc_ci_bootstrap(y_test, y_proba)

    # ==============================
    # 2️⃣ Determine Optimal Cutoff (Youden Index)
    # ==============================
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)

    youden_index = tpr - fpr
    optimal_idx = np.argmax(youden_index)
    optimal_threshold = thresholds[optimal_idx]

    # ==============================
    # 3️⃣ Classification Using Youden Cutoff
    # ==============================
    y_pred = (y_proba >= optimal_threshold).astype(int)

    # ==============================
    # 4️⃣ Metrics
    # ==============================
    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    cm = confusion_matrix(y_test, y_pred)
    tn, fp, fn, tp = cm.ravel()

    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    youden_final = sensitivity + specificity - 1

    results.append([
        name, optimal_threshold,
        auc, lower_ci, upper_ci,
        acc, precision, recall, f1,
        sensitivity, specificity, youden_final
    ])

performance_df = pd.DataFrame(results, columns=[
    "Model", "Optimal Cutoff",
    "AUC", "AUC 2.5%", "AUC 97.5%",
    "Accuracy", "Precision", "Recall", "F1",
    "Sensitivity", "Specificity", "Youden Index"
])

performance_df = performance_df.sort_values("AUC", ascending=False)

print("\n📊 MODEL PERFORMANCE (Youden Cutoff)")
print("="*90)
print(performance_df)
```

```python
# ==========================================================
# 🌍 RANDOM FOREST — COUNTRY-WISE PERFORMANCE (TEST DATA)
# ==========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import (
    roc_auc_score, roc_curve,
    accuracy_score, precision_score,
    recall_score, f1_score,
    confusion_matrix
)

# ==========================================================
# 1️⃣ RESET INDEX TO AVOID ALIGNMENT ERROR
# ==========================================================

X_test_country = X_test_consensus_df.reset_index(drop=True)
y_test_country = pd.Series(y_test).reset_index(drop=True)
y_proba_rf_country = pd.Series(y_pred_proba_rf).reset_index(drop=True)

country_cols = [col for col in X_test_country.columns if col.startswith("country_")]

# ==========================================================
# 2️⃣ BOOTSTRAP AUC 95% CI FUNCTION
# ==========================================================

def auc_ci_bootstrap(y_true, y_proba, n_bootstraps=2000):
    
    rng = np.random.RandomState(42)
    scores = []
    
    y_true = np.array(y_true)
    y_proba = np.array(y_proba)
    
    for _ in range(n_bootstraps):
        idx = rng.randint(0, len(y_proba), len(y_proba))
        
        if len(np.unique(y_true[idx])) < 2:
            continue
        
        score = roc_auc_score(y_true[idx], y_proba[idx])
        scores.append(score)
    
    lower = np.percentile(scores, 2.5)
    upper = np.percentile(scores, 97.5)
    
    return lower, upper

# ==========================================================
# 3️⃣ FIND BEST THRESHOLD BY MAXIMIZING F1 SCORE
# ==========================================================

def find_best_f1_threshold(y_true, y_proba):
    
    thresholds = np.linspace(0.01, 0.99, 200)
    best_f1 = 0
    best_thresh = 0.5
    
    for t in thresholds:
        y_pred = (y_proba >= t).astype(int)
        f1 = f1_score(y_true, y_pred, zero_division=0)
        
        if f1 > best_f1:
            best_f1 = f1
            best_thresh = t
            
    return best_thresh, best_f1

# ==========================================================
# 4️⃣ COUNTRY-WISE EVALUATION
# ==========================================================

results = []

plt.figure(figsize=(10,8))

for country in country_cols:
    
    idx = X_test_country[country] == 1
    
    if idx.sum() < 30:   # Skip very small samples
        continue
    
    y_true_c = y_test_country[idx]
    y_proba_c = y_proba_rf_country[idx]
    
    # -----------------------------
    # AUC + 95% CI
    # -----------------------------
    if len(np.unique(y_true_c)) > 1:
        auc = roc_auc_score(y_true_c, y_proba_c)
        lower_ci, upper_ci = auc_ci_bootstrap(y_true_c, y_proba_c)
    else:
        auc = np.nan
        lower_ci, upper_ci = np.nan, np.nan
    
    # -----------------------------
    # F1-Optimized Threshold
    # -----------------------------
    best_thresh, best_f1 = find_best_f1_threshold(y_true_c, y_proba_c)
    y_pred_c = (y_proba_c >= best_thresh).astype(int)
    
    # -----------------------------
    # Metrics
    # -----------------------------
    acc = accuracy_score(y_true_c, y_pred_c)
    precision = precision_score(y_true_c, y_pred_c, zero_division=0)
    recall = recall_score(y_true_c, y_pred_c, zero_division=0)
    
    cm = confusion_matrix(y_true_c, y_pred_c)
    tn, fp, fn, tp = cm.ravel()
    
    sensitivity = tp/(tp+fn) if (tp+fn)>0 else 0
    specificity = tn/(tn+fp) if (tn+fp)>0 else 0
    
    results.append([
        country.replace("country_","").capitalize(),
        len(y_true_c),
        auc,
        lower_ci,
        upper_ci,
        best_thresh,
        acc,
        precision,
        recall,
        best_f1,
        sensitivity,
        specificity
    ])
    
    # -----------------------------
    # ROC Curve
    # -----------------------------
    fpr, tpr, _ = roc_curve(y_true_c, y_proba_c)
    
    plt.plot(
        fpr, tpr,
        label=f"{country.replace('country_','').capitalize()} "
              f"(AUC={auc:.3f} [{lower_ci:.3f}-{upper_ci:.3f}])"
    )

# Diagonal reference
plt.plot([0,1],[0,1],'k--', linewidth=1)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Random Forest ROC Curve by Country (Test Data)")
plt.legend(fontsize=8)
plt.grid(alpha=0.3)

# 🔥 Save as TIFF
plt.savefig(
    "Random_Forest_ROC_by_Country.tiff",
    format="tiff",
    dpi=600,
    bbox_inches="tight"
)

plt.show()
# ==========================================================
# 5️⃣ FINAL SUMMARY TABLE
# ==========================================================

country_perf_df = pd.DataFrame(results, columns=[
    "Country",
    "N (Test)",
    "AUC",
    "AUC 2.5%",
    "AUC 97.5%",
    "Best F1 Cutoff",
    "Accuracy",
    "Precision",
    "Recall",
    "F1 Score",
    "Sensitivity",
    "Specificity"])

country_perf_df = country_perf_df.sort_values("AUC", ascending=False)

print("\n📊 RANDOM FOREST PERFORMANCE BY COUNTRY (F1-Optimized)")
print("="*110)
print(country_perf_df.round(3))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import roc_curve
import joblib
# Example: if your encoded test data came from
X_test_raw = X_test.copy()  # X_test has raw string columns before any encoding
# ---------------------------
# 1️⃣ Define categorical columns
# ---------------------------
categorical_cols = [
    'matage','placeresid','latrine','famsize','num5child','sexofhead',
    'litracy','bottlefeeding','disposalofchild','maritalstatus','working',
    'twin','childgender','childage','birthinterval','breastfed','anc',
    'birthsize','diarrhea','fever','vitamina','maternaldecision','dietdivers',
    'watersource','ari','country'
]

# If numeric columns exist, you can add them here
numeric_cols = []  # e.g., child_weight, maternal_height

# ---------------------------
# 2️⃣ Preprocessing
# ---------------------------
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
        # ('num', StandardScaler(), numeric_cols)  # uncomment if numeric_cols exist
    ]
)

# ---------------------------
# 3️⃣ Random Forest Classifier
# ---------------------------
rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=15,
    min_samples_leaf=5,
    random_state=42,
    n_jobs=-1
)

# ---------------------------
# 4️⃣ Create pipeline
# ---------------------------
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('rf', rf_model)
])

# ---------------------------
# 5️⃣ Fit pipeline on raw training data
# ---------------------------
pipeline.fit(X_train_bal_raw, y_train_bal)

# ---------------------------
# 6️⃣ Predict probabilities on raw test data
# ---------------------------
y_pred_proba = pipeline.predict_proba(X_test_raw)[:,1]

# ---------------------------
# 7️⃣ Calculate optimal threshold (Youden Index)
# ---------------------------
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
youden = tpr - fpr
optimal_idx = youden.argmax()
optimal_threshold = thresholds[optimal_idx]

print(f"✅ Optimal Youden threshold: {optimal_threshold:.3f}")

# ---------------------------
# 8️⃣ Save pipeline and threshold
# ---------------------------
joblib.dump(pipeline, "malnutrition_rf_pipeline.pkl", compress=3)
joblib.dump(optimal_threshold, "youden_threshold.pkl")

print("✅ Compact production model saved")
```
